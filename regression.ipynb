#Linear Regression

Using linear regression to predict heights based on weights.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error

# examples of values
weights = np.array([60, 65, 72, 75, 78, 82, 85, 90, 95, 100])
heights = np.array([160, 165, 170, 172, 175, 173, 178, 180, 183, 185])

# creates and trains the model
model = LinearRegression()
model.fit(weights.reshape(-1, 1), heights)

# array with a range of weights to generate the line
line_weights = np.array([50, 110])

# calculating the mean absolute error
heights_predicts = model.predict(weights.reshape(-1, 1))
mae = mean_absolute_error(heights, heights_predicts)
print(f"MAE = {mae:.2f} cm")

# predicts the heights corresponding to the weights
line_heights = model.predict(line_weights.reshape(-1, 1))

# plotting the graph
plt.scatter(weights, heights, color='blue')
plt.plot(line_weights, line_heights, color='red')
plt.title('Linear Regression: Weight → Height')
plt.xlabel('Weight (kg)')
plt.ylabel('Height (cm)')
plt.grid(True)
plt.show()
```

Linear Regression Results:

The graph shows a positive linear trend between weight and height with a mean absolute error of 1.04 cm. The blue dots represent the observed data, and the red line represents the fitted model.



#Logistic Regression

Using logistic regression to predict the chance of a student passing or failing based on the number of hours of study.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# examples - (0=failed, 1=approved)
study_hours = np.array([1, 2, 2.5, 3, 4, 4.5, 5, 6, 7, 8, 9, 10])
results = np.array([0, 0, 0, 0, 0, 1, 0, 1, 1, 1, 1, 1])

# creates and trains the model
model = LogisticRegression()
model.fit(study_hours.reshape(-1, 1), results)

# making predictions with the model
predicts = model.predict(study_hours.reshape(-1, 1))

# calculating accuracy
accuracy = accuracy_score(results, predicts)
print(f"Accuracy: {accuracy:.2f}")

# predicting the odds of approval
hours_line = np.linspace(0, 12, 100)
probabilidades = model.predict_proba(hours_line.reshape(-1, 1))[:, 1]

# plotting the graph
plt.scatter(study_hours[results==1], np.ones(sum(results==1)),
           color='green', label='Approved')
plt.scatter(study_hours[results==0], np.zeros(sum(results==0)),
           color='red', label='Failed')
plt.plot(hours_line, probabilidades, label='Probability of approval')
plt.title('Logistic Regression: Study Hours → Probability of approval')
plt.xlabel('Study Hours')
plt.ylabel('PProbability of approval')
plt.yticks([0, 0.5, 1], ['0%', '50%', '100%'])
plt.grid(True)
plt.legend()
plt.show()
```

Logistic regression results:

The graph shows red dots representing failed students, green dots representing passed students, and a passing probability curve with an accuracy of 0.83.
